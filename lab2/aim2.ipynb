# Bitwise Logical Operations on Binary Images

This notebook demonstrates fundamental bitwise logical operations—**OR**, **AND**, **NOT**, **XOR**, and **XNOR**—applied to thresholded medical images (CT and MRI).

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Image Preprocessing and Binary Conversion

```python
img1 = cv2.imread('ct.png', 0)
img2 = cv2.imread('mri.png', 0)

if img1 is not None and img2 is not None:
    img2 = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

    def custom_bin(img, thresh=127):
        return np.where(img > thresh, 255, 0).astype(np.uint8)

    b1 = custom_bin(img1)
    b2 = custom_bin(img2)
else:
    print("Error: Image files not found.")
```

### Applying Bitwise Operations

```python
custom_or = b1 | b2
custom_and = b1 & b2
custom_not = 255 - b1
custom_xor = b1 ^ b2
custom_xnor = 255 - custom_xor
```

### Visualization
Note: `cv2.imshow` is replaced with `matplotlib` for inline notebook viewing.

```python
titles = ['Binary CT', 'Binary MRI', 'OR', 'AND', 'NOT (CT)', 'XOR', 'XNOR']
images = [b1, b2, custom_or, custom_and, custom_not, custom_xor, custom_xnor]

plt.figure(figsize=(20, 10))
for i in range(len(images)):
    plt.subplot(1, 7, i + 1)
    plt.imshow(images[i], cmap='gray')
    plt.title(titles[i])
    plt.axis('off')

plt.tight_layout()
plt.show()
```
